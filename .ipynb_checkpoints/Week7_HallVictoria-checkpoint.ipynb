### Part 1 PCA and Variance Threshold in Linear Regression

```python
#loading libraries
import pandas as pd
```

#### Step 1

```python

housing_df = pd.read_csv("housing_train.csv")
housing_df.head(10)
housing_df.shape
```

#### Step 2

```python
#Dropping id

#Checking col names
housing_df.columns
    

#dropping id column
housing_df = housing_df.drop(columns=['Id'])

```

```python
#Looking for columns with missing values
percent_missing = housing_df.isnull().sum()*100/len(housing_df)

#Making a df of the missing %
missing_df = pd.DataFrame({'column_name':housing_df.columns,'percent_missing': percent_missing})

#Subsetting to only include those columns missing more than 40%
remove_col = missing_df[missing_df['percent_missing'] >= 40]

print("There are {} columns missing more than 40% of data".format(remove_col.shape[0]))
```

```python
#Removing features missing 40% or more of data
remove_col

housing_df = housing_df.drop(columns=['Alley','FireplaceQu','PoolQC','Fence','MiscFeature'])
```

#### Step 3 & 4

```python
#Replacing missing numeric values with median for each col and categorical with mode

#Getting column types
housing_numeric = housing_df[housing_df.columns[housing_df.dtypes != 'object']]
housing_cat = housing_df[housing_df.columns[housing_df.dtypes == 'object']]

#Finding median of numeric types
housing_medians = housing_numeric.median()

#Finding mode of categorical data
housing_modes = housing_cat.mode().iloc[0]

#replacing miss numeric values with median
housing_numeric.fillna(housing_medians, inplace = True)

#replacing miss categorical values with mode
housing_cat.fillna(housing_modes, inplace = True)

housing_df = pd.concat([housing_numeric,housing_cat],axis=1)
```

```python
#Checking the above steps worked
housing_df.isnull().sum().sort_values(ascending=False)
```

#### Step 5

```python
#Creating dummy values for categorical data. Get dummies automaticlly applies to categorical data only and drops the original column after complete. 
housing_df = pd.get_dummies(housing_df)

housing_df.head(5)
```

#### Step 6

```python
#Spliting the data into train and test
#loading Package
from sklearn.model_selection import train_test_split
import numpy as np


#Creating features and target
x = housing_df.drop('SalePrice',1)
y = housing_df['SalePrice']

#Creating splits with 20% of data for the test
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=.2)

```

#### Step 7

```python
#Importing packages
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn import metrics


#Creating model object and applying to data
olsmodel=LinearRegression()
olsmodel.fit(x_train,y_train)

#Running model on test data
y_pred = olsmodel.predict(x_test)

#RMSE
print('RMSE of Test Data:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))

#R2
print('R2 of Test Data:',metrics.r2_score(y_test,y_pred))
```

#### Steps 8 and 9

```python
#loading libraries
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#Standardizing the features
features = StandardScaler().fit_transform(x_train)

#Creating PCA that retains 90% of the variance
pca=PCA(n_components=0.9, whiten=True)

#Conducing PCA
features_pca = pca.fit_transform(features)

print("There were {} features originally".format(features.shape[1]))
print("There are now {} features after conducting PCA".format(features_pca.shape[1]))
```

#### Steps 10 and 11

```python
#Scaling features of test data
sc = StandardScaler()
features = sc.fit_transform(x_train)
test_features = sc.transform(x_test)


#Conducting PCA
x_test_pca = pca.transform(test_features)

#Fitting new model
olsmodel.fit(features_pca, y_train)

#Using model to predict on test
model_pred_pca = olsmodel.predict(x_test_pca)

#RMSE
print('RMSE of Test Data:',np.sqrt(metrics.mean_squared_error(y_test,model_pred_pca)))

#R2
print('R2 of Test Data:',metrics.r2_score(y_test,model_pred_pca))

```

#### Steps 12-15

```python
#importing library
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold

#Applying minmax scaler to features, fitting to train, only transforming on test
minmax = MinMaxScaler()
x_train_minmax = minmax.fit_transform(x_train)
x_test_minmax = minmax.transform(x_test)

#Creating thresholder
thresholder = VarianceThreshold(threshold=0.1)

#apply threshold to minmax scaled features
x_train_highvariance = thresholder.fit_transform(x_train_minmax)
x_test_highvariance = thresholder.transform(x_test_minmax)

#Fitting new model
olsmodel.fit(x_train_highvariance, y_train)

#Using model to predict on test
model_pred_minmax = olsmodel.predict(x_test_highvariance)

#RMSE
print('RMSE of Test Data:',np.sqrt(metrics.mean_squared_error(y_test,model_pred_minmax)))

#R2
print('R2 of Test Data:',metrics.r2_score(y_test,model_pred_minmax))
```

#### Step 16
As we reduced the dimensionality of our data features we saw a reduction in our R2. This makes sense because as we remove features, we will account for less and less of the varaince in our data. This is a common trade off with feature reduction. It also difficult to determine what the new features are because after these transformation they are not interpretable to humans. . However, reducing the dimensionality does reduce the time and storage space by getting rid of redundancies. In this same vein, it also helps eliminate multicollinearity. 

### Part 2 Categorical Feature Selection

#### Steps 1-3 

```python
#importing data frame
mushroom_df = pd.read_csv("mushrooms.csv")
print(mushroom_df.shape)
mushroom_df.head(5)
```

```python
#checking class info
mushroom_df['class'].value_counts()

#Creating dummy variables for features
mushroom_df = pd.get_dummies(mushroom_df, columns=mushroom_df.columns[1:])
```

```python
#Splitting into training and test set
#Creating features and target
x = mushroom_df.drop('class',1)
y = mushroom_df['class']

#Creating splits with 20% of data for the test
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=.2)
```

#### Step 4 and 5

```python
#importing decision tree
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)


```

```python
#importing functions for accuracy and conf matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report



clf_pred = clf.predict(x_test)

print("The Accuracy score of the logistic regression model is {:.2%}:".format(accuracy_score(y_test,clf_pred)))



#Creating confusion matrix
cf_matrix = confusion_matrix(y_test,clf_pred)
#Using seaborn to better visualize the matrix
ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
ax.set_title('Mushroom Confusion Matrix \n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])
plt.show()
```

#### Step 6

```python
#Creating visual of the decision tree
fig = plt.figure(figsize=(25,20))
tree.plot_tree(clf,filled=True)
```

#### Steps 7 and 8

```python
#importing libraries
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_classif


#Creating selector
chi2_selector = SelectKBest(score_func=chi2, k=5)

#applying
features_kbest = chi2_selector.fit_transform(x_train,y_train)

#identifying top 5 features
cols = chi2_selector.get_support(indices=True)
featurs_df_new = x_train.iloc[:,cols]
featurs_df_new

print("Top 5 Features Selected")
print("------------------------")
for col in featurs_df_new.columns:
    print(col)
```

#### Step 9

```python

#Creating model only using selected features
clf = tree.DecisionTreeClassifier()
clf = clf.fit(features_kbest,y_train)

x_test1 = x_test[['odor_f','odor_n','gill-color_b','stalk-surface-above-ring_k','stalk-surface-below-ring_k']]


clf_pred = clf.predict(x_test1)

print("The Accuracy score of the logistic regression model is {:.2%}:".format(accuracy_score(y_test,clf_pred)))



#Creating confusion matrix
cf_matrix = confusion_matrix(y_test,clf_pred)
#Using seaborn to better visualize the matrix
ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
ax.set_title('Selected Features Mushroom Confusion Matrix \n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])
plt.show()
```

#### Step 10
Our model was slightly less accurate when we only accounted for the top 5 features based on their Chi2 values. This makes sense as we have reduced the dimensionality of our model, meaning there are features that were previously accounting for variance that are no longer in the model. Still, the model is highly accurate with the exception of 7% of our predictions False negatives.
