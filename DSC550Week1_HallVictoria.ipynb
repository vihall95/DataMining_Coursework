###  First five games in the data frame on the SNES platform

```python
#Using groupby and get_group from PD to find the first 5 games on SNES platform
snesdf = df_videogames.groupby("Platform")
snesdf.get_group("SNES").head(5)
```

### Find the five publishers with the highest total global sales

```python
# Using groupby and summing the global sales. Then sorted values by top 5
publisherval = df_videogames.groupby("Publisher")["Global_Sales"].sum()
publisherval.sort_values(axis = 0, ascending = False).head(5)
```

### New Column with percentage of global sales 

```python
#First, inserting new column that is North American Values as percentage of global sales. Then converting to a percentage and finally show first 5 values
df_videogames["American_Vs_Global"] = df_videogames.NA_Sales/df_videogames.Global_Sales
df_videogames["American_Vs_Global"] = pd.Series(["{0:.2f}%".format(val * 100) for val in df_videogames["American_Vs_Global"]], index = df_videogames.index)
df_videogames.head(5)
```

### Finding NaN values
