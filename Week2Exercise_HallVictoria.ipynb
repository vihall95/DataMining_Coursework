### The dataset

The dataset is provided by Kaggle. It is a dataset of performance grades of exercises. It includes data like height, weight, body fat, as well as number of sit-ups, grip-force, etc. 

https://www.kaggle.com/kukuroo3/body-performance-data

###  Question

Is there a relationship between different body measurements like height, weight, and body fat to the number of situps performed. 

```python
#Importing libraries and reading in file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
body_df = pd.read_csv("bodyPerformance.csv")
```

```python
#Exploring the dataframe
body_df.head(10)
len(body_df)
body_df.isnull().sum()
```

### Scatter Plots

Below are two scatter plots. One is height vs number of situps and the other is body fat versus number of situps. I added jitter and changed the transparency to help with some of the over plotting. In both instances it does appear like there is a relationship between these body metrics and number of situps completed. We see heavy concentrations with small outliers which tells us that we may be working with a normal distribution but further visualizations and tests need to be performed. For height, there does appear to be a positive relationship meaning the taller the participant the more situps they complete. The relationship is inverse for body fat where we see the that the higher the body fat, less situps.

```python
#Creating scatter plots
body_df.head(5)
list(body_df.columns)
#Creating variables with columns of interest and changing names
sit_ups = body_df['sit-ups counts']
height = body_df['height_cm']
weight = body_df['weight_kg']
body_fat = body_df['body fat_%']

#First scatter plot chart. Adding jitter to help with some of the overplotting.
plt.figure(1)
sb.regplot(x=height,y=sit_ups, fit_reg = False, x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha': 1/5})
plt.title("Height vs Number of Situps Completed")
plt.xlabel("Height (CM)")
plt.ylabel("Number of Situps")


#Second plot
plt.figure(2)
sb.regplot(x=body_fat,y=sit_ups, fit_reg = False, x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha': 1/5})
plt.title("Body Fat vs Number of Situps Completed")
plt.xlabel("Body Fat (%)")
plt.ylabel("Number of Situps")
```

### Histograms

The below charts are histograms of each of our variables of interest. I've also included  kernel density estimations (IE estimating the PDF of situps) to show us a smooth curve of the data. This chart confirms suspicions that the distribution for this data is normal. Knowing this points in the direction of what models we can use. Of course we would need to test further, but this demonstrates that we probably aren't violating one our parametric assumptions. 

```python
plt.figure(3)
sb.histplot(x=sit_ups,kde=True)
plt.title('Situps Histogram')
plt.xlabel('Number of Sit Ups')
plt.ylabel('# of Invidivuals')
```

```python
plt.figure(4)
sb.histplot(x=height,kde=True)
plt.title('Height Histogram')
plt.xlabel('Height (CM)')
plt.ylabel('# of Invidivuals')


plt.figure(5)
sb.histplot(x=body_fat,kde=True)
plt.title('Body Fat Histogram')
plt.xlabel('Body Fat (%)')
plt.ylabel('# of Invidivuals')
```

### Lineplot

The lineplots below further explore the relationships between our variables. These confirm the relationship we saw earlier in the scatter plots but they emphasize some interesting activity in the tail ends of our distributions suggesting we need to further review outliers and possibly modify the data we are using. 

```python
plt.figure(6)
sb.lineplot(x=body_fat,y=sit_ups)
plt.title("Body Fat vs Number of Situps")
plt.xlabel("Body Fat %")
plt.ylabel("Number of Sit Ups")
```

```python
plt.figure(5)
sb.lineplot(x=height,y=sit_ups)
plt.title("Height vs Number of Situps")
plt.xlabel("Height (CM)")
plt.ylabel("Number of Sit Ups")
```

### Conclusion

We created three types of visualizations for our data. First we created simple scatter plots to see if there is any relationship between our independent variables of body fat and height to our dependent variable of number of situps completed. In both of these exercises we saw that there did appear to be a relationship where lower body fat meant more situps and taller individuals completed more. Second, we created histograms of our variables of interest and estimates of the PDF to see if our data was normally distributed. In all instances it does appear to be more normally distributed, confirming one of our assumptions about our data. Finally, out linecharts further demonstrated the relationship between our variables but also showed us some abnormal activity in the outliers of our data, letting us we need to explore further and make sure these data points are real and not errors as well as figure out how to handle these.
